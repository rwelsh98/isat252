```python
import pandas
```

```python
df=pandas.read_excel('s3://isat252.welshrj/house_price.xls')
df.plot(x='built_in',y='price')
```

## line chart

# avg prices per year

```python
avg_price_per_yr=df.groupby('built_in').mean()['price']
avg_price_per_yr.plot()
```

# Histogram

## to show the dist. of prices

```python
df['price'].hist()
```

most house prices are around $200,000.

## Bar chart

# avg prices per house type

```python
avg_price_per_type=df.grouby('house_type').mean()['price']

avg_price_per_type.plot.bar() #return a bar chart
```

single family houses have the highest average price

# Pie Chart

## the number of record per house type

```python
count_per_house_type=df['house_type'].value_counts()
count_per_house_type.plot.pie()
```

# Scatter Plot

## relationship between price and area

```python
df.plot.scatter(x='area',y='price')
```
