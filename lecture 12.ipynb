## Linear regression

```python
from scipy import stats
```

```python
result=stats.linregress(df['WEIGHT'],df["PRICE"]) #build a SLR, x first, y second
```

```python
print('The slope is {}'.format(result.slope))
print('The intercept is {}'.format(result.intercept))
print('THe r square is {}'.format(result.rvalue*result.rvalue))
print('The std error is {}'.format(result.stderr))
```

```python
print('The price of {} weight diamond is ${}.'.format(0.9,0.9*result.slope+result.intercept))
```
